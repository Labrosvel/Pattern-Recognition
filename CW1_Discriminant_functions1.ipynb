# Vector Calculations

```python
import numpy as np
at= np.array([1,0,0]) #weights
bt = np.array([1,0.5,2.5,2.5,0.5,1]) #margin
yt = np.array([[1,0,2], [1,1,2], [1,2,1], [-1,3,-1], [-1,2,1], [-1,3,2]]) #features
n = 0.1
```

```python
for i in range(0,6):
    print(np.inner(at,yt[i,:]))
    at = at+n*(bt[i]-np.inner(at,yt[i,:]))*yt[i,:]
    print(at)
```
