```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
xt = iris.data
#xt
#xt.shape
```

```python
yt_temp = np.c_[np.ones(xt.shape[0]), xt]
#yt_temp[:51,:]
```

```python
iris.target # Explore the data
#iris.target.shape
```

```python
for i in range(0,150):
    if iris.target[i]==0:
        yt_temp[i,:] = yt_temp[i,:]*1
    else:
        yt_temp[i,:] = yt_temp[i,:]*(-1)
yt = yt_temp
yt
```
